# Redes neuronales

Son unidades artificiales que, interrelacionadas entre sí, componen un modelo o sistema matemático que intenta simular el funcionamiento del cerebro humano, es decir, un sistema inteligente capaz de aprender a realizar ciertas tareas.

Una red neuronal está constituida por distintas capas de neuronas con diferentes funciones que van aprendiendo de la información que les llega y asumiendo generalizaciones acerca de esta (método de aprendizaje). Una neurona recibe información de entrada, la procesa y produce una información de salida.

Una red neuronal se compone de "neuronas" organizadas en capas. Cada neurona de la red (exceptuando la capa de entrada) es en realidad un sumatorio de todas sus entradas; que no son más que las salidas de las capas anteriores multiplicadas por unos pesos. A esta suma se le añade un término adicional llamado sesgo o bias. Y al resultado se le aplica una función no lineal conocida como función de activación.



Una neurona artificial está formada por los siguientes componentes:

1. **Entradas:** Las redes neuronales reciben datos de entrada, representados como un vector $x = (x_1, x_2, \ldots, x_n)$, donde $n$ es el número de características o atributos de entrada.

2. **Pesos:** Cada conexión entre una neurona de una capa y una neurona de la siguiente capa tiene asociado un peso $w$, que representa la importancia o contribución de la neurona de origen a la neurona de destino.

3. **Función de activación:** Después de que los datos de entrada se ponderan por los pesos, se aplica una función de activación $f$ a la suma ponderada de las entradas y los pesos más un término de sesgo (bias), representado como $z = \sum_{i=1}^{n} w_i x_i + b$, donde $b$ es el sesgo.

4. **Salida:** La salida de la función de activación se convierte en la entrada de las neuronas de la siguiente capa, y el proceso se repite a través de todas las capas de la red hasta llegar a la capa de salida.

5. **Aprendizaje:** Durante el entrenamiento, la red neuronal ajusta los pesos y sesgos para minimizar una función de pérdida, que mide la diferencia entre las salidas predichas y las salidas reales. Esto se hace típicamente utilizando algoritmos de optimización como el descenso del gradiente.

# Aprendizaje de las redes neuronales mediante descenso del gradiente

Ante la expansión y generalización de la IA en todo el ecosistema tecnológico, las redes neuronales adquieren una gran importancia, y uno de sus principales algoritmos de entrenamiento es el de gradient descent.

El algoritmo de gradiente descendente es un mecanismo de entrenamiento de sistemas de aprendizaje automático como los basados en redes neuronales. Es el más extendido y utilizado para el aprendizaje o entrenamiento de redes neuronales debido a su sencillez y facilidad de implementación.

El aprendizaje o entrenamiento de las redes neuronales consiste en determinar los pesos sinápticos de las redes neuronales que permitan aprender a la red y así tenga la capacidad de ejecutar la tarea en cuestión, partiendo de valores aleatorios de estos. Matemáticamente, radica en ajustar los pesos que minimizan una función de error acerca del rendimiento de la red, y aquí es donde juega un papel fundamental el algoritmo de gradiente descendiente.

Este algoritmo permite encontrar los parámetros que minimizan una función matemática, conocida como ‘función de coste’, que en el caso de las redes neuronales es la función que calcula el error que comete la red en la ejecución de la tarea. Para el aprendizaje de la red, se parte de un conjunto aleatorio de pesos y a medida que se van ejecutando iteraciones del algoritmo, los pesos se van ajustando y el error de la red es menor.

El algoritmo de gradiente descendente es el método de aprendizaje más utilizado para entrenar redes neuronales, ya que además de su sencillez y facilidad de implantación, permite la retroalimentación o propagación hacia atrás del error que comete la red neuronal en la ejecución de su tarea, posibilitando que todas las neuronas conozcan su contribución al error en función de sus pesos sinápticos y los vayan ajustando.

Las redes neuronales son modelos de IA que consisten en capas de nodos interconectados, o neuronas, que realizan transformaciones no lineales en los datos de entrada. La salida de la red depende de los pesos y sesgos de las neuronas, que son los parámetros que deben optimizarse. La función de coste mide la eficacia con la que la red predice el resultado deseado, como las etiquetas de una tarea de clasificación o los valores de una tarea de regresión. Mediante el uso del descenso de gradiente, la red puede ajustar sus parámetros para reducir la función de coste y mejorar su rendimiento.

### Descenso de gradiente

El descenso de gradiente es un algoritmo iterativo que intenta minimizar una función determinada, generalmente llamada función de costo o pérdida, actualizando sus parámetros en función del gradiente de la función. El gradiente es un vector que apunta en la dirección del mayor aumento de la función, por lo que moverse en la dirección opuesta reduce el valor de la función. El algoritmo comienza con una estimación inicial de los parámetros, y luego los actualiza restando una fracción del gradiente, llamada tasa de aprendizaje, hasta que alcanza un mínimo local o un criterio de convergencia.

Descenso del gradiente (estocástico) en mini-lotes (o mini-batch): en lugar de alimentar la red con una única muestra, se introducen N muestras en cada iteración.

Las iteraciones que lleva a cabo el algoritmo del Descenso del Gradiente:

1. Introducimos un mini-lote de entrada con N muestras aleatorias, previamente etiquetado (lo que significa que conocemos la salida real).

2. Se inicializan aleatoriamente los pesos y los sesgos de la red neuronal. rompiendo su simetría. El fin es evitar que todas las neuronas de una capa acaben aprendiendo lo mismo.

3. Después de los cálculos pertinentes en cada capa de la red, obtenemos como resultado las predicciones a su salida.

4. Evaluamos la función de coste (también llamada función de pérdida) para dicho mini-lote. Se trata de una función elegida previamente en base al tipo de problema concreto, para poder evaluar de la forma más adecuada la diferencia entre las predicciones de nuestra red y las salidas reales. El valor de esta función de coste es lo que se trata de minimizar en todo momento mediante el algoritmo, y hacia ello se orientan los siguientes pasos.

5. Calculamos el gradiente los parámetros de la red. Gráficamente sería la pendiente de la tangente a la función de coste en el punto donde nos encontramos (evaluando los pesos actuales). Matemáticamente es un vector que nos da la dirección y el sentido en que dicha función aumenta más rápido, por lo que deberíamos movernos en sentido contrario si lo que tratamos es de minimizarla.

6. Una vez obtenido el vector gradiente, actualizaremos los parámetros de la red restando a su valor actual el valor del gradiente correspondiente, multiplicado por una tasa de aprendizaje que nos permite ajustar la magnitud de nuestros pasos. El término de actualización se resta porque como ya hemos visto antes queremos avanzar en el sentido contrario al del gradiente, para que la función de coste disminuya.
$$
   w_{nuevo} = w_{viejo} - \alpha \frac{\partial L}{\partial w}
$$
   Donde $w_{nuevo}$ es el nuevo peso, $w_{viejo}$ es el peso anterior, $\alpha$ es la tasa de aprendizaje que controla qué tan grandes son los pasos de actualización y $\frac{\partial L}{\partial w}$ es la derivada parcial de la función de pérdida con respecto al peso.
   
Repetiremos todos los pasos mientras el valor de la función de coste y las métricas de salida no empiecen a empeorar de forma sostenida.

El descenso de gradiente es esencial para el entrenamiento de redes neuronales porque permite ajustar los pesos y sesgos de manera iterativa para que la red pueda aprender a hacer predicciones más precisas y minimizar el error en sus salidas. La tasa de aprendizaje $\alpha$ es un hiperparámetro crítico que debe ajustarse para garantizar una convergencia eficiente y evitar problemas como el sobreajuste o el subajuste.

```python
#Imports
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import math
%matplotlib inline
%config InlineBackend.figure_formats = ['svg']
np.random.seed(seed=13)
```

```python
h=1e-6
#Definición de la función
def fun(x):
    f= (x)**2 +2
    return f
#Derivada de la función
def prima(x):
    prima= 2*x
    return prima
#Derivación automática central
def funpc(x):
    fprima=(fun(x+h)- fun(x-h))/(2*h)
    return fprima
#Error
def err(x):
    error=prima(x) - funpc(x)
    return error
#Punto dado para calcular la derivada
x0=7
#Calculo de la derivada en el punto dado
df_result = prima(x0)
print("El valor de la derivada (prima) en x =", x0, "es:", df_result)
df_result = funpc(x0)
print("El valor de la derivada (central) en x =", x0, "es:", df_result)
```

```python
#Graficar
x=np.linspace(-7,7)
g=fun(x)
p=prima(x)
c=funpc(x)
e=err(x)
plt.plot(x,g,color='r')
plt.plot(x,p,color='b')
plt.plot(x,c,color='y')
plt.plot(x,e)
plt.legend(['f(x)','Derivada','DA central','Error'])
plt.grid(True)
plt.show()
```

```python
#Descenso de gradiente con diferenciación automática
#Parámetros del descenso de gradiente
alpha=0.9  #Tasa de aprendizaje
itera=10  #Número de iteraciones
valor=0  #Valor inicial

#Descenso de gradiente
def descengad(x):
    for i in range(itera):
        gradient=funpc(x)  #Calcular el gradiente en el punto actual
        x=x-(alpha*gradient)  #Actualizar el valor de x
    return x
GD=descengad(valor)
#Imprimir el resultado
print("El mínimo de la función se encuentra en x =", GD)
print("El valor mínimo de la función es f(x) =", fun(GD))
```

```python
#Definir el vector de entradas muestras de x
x=np.random.uniform(0, 1, 30)
#Generar los valores t objetivos
#Definir la función que representa la linea que genera t sin ruido
def f(x):
    return x*7
#Generar los objetivos t con ruido
ruido_v=0.2  #Ruido gaussiano
#Error de ruido para cada muestra de x
ruido=np.random.randn(x.shape[0])*ruido_v
t=f(x)+ruido
```

```python
#Graficar los valores objetivos t contra las entradas x
plt.plot(x,t,'o',label='t')
#Graficar f(x)
ff=f(x)
plt.plot(x, ff, label='f(x)')
plt.xlabel('x', fontsize=12)
plt.ylabel('t', fontsize=12)
plt.title('Entradas x vs Objetivos t')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Definir de salida y
def outy(x,w):
    return x*w
#Definir la función de perdida (MSE)
def loss(x,w,t):
    return np.mean((t-outy(x,w))**2)

#Graficar la función de perdida contra el peso w dado
#Vector de pesos para poder graficar contra la perdida
ws=np.linspace(0,4, num=100)
#Perdida para cada peso
loss_ws=np.vectorize(lambda w:loss(x,w,t))(ws)
#Graficar
plt.plot(ws,loss_ws,label='Perdida')
plt.xlabel('w', fontsize=12)
plt.ylabel('$\\xi$', fontsize=12)
plt.title('Función de perdida respecto a w')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Derivada de la función
def primal(x,w,t):
    return np.mean(2*x*(outy(x,w)-t))
#Derivación automática central
def funpcl(x,w,t):
    return (loss(x,w+h,t)-loss(x,w-h,t))/(2*h)
#Error respecto a la derivación central
def errl(x,w,t):
    return primal(x,w,t) - funpcl(x,w,t)

#Graficar la función de peridad contra el peso w dado
#Vector de pesos para poder graficar
ws=np.linspace(0, 4, num=100)
loss_ws=np.vectorize(lambda w:loss(x,w,t))(ws)
loss_p=np.vectorize(lambda w:primal(x,w,t))(ws)
loss_c=np.vectorize(lambda w:funpcl(x,w,t))(ws)
loss_e=np.vectorize(lambda w:errl(x,w,t))(ws)

#Graficar
plt.plot(ws,loss_ws,label='Función de perdida')
plt.plot(ws,loss_p,color='r',label='Derivada')
plt.plot(ws,loss_c,color='y',label='DA central')
plt.plot(ws,loss_e,color='b',label='Error')
plt.xlabel('w',fontsize=12)
plt.ylabel('$\\xi$',fontsize=12)
plt.title('Función de perdida con su derivada respecto a w')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Descenso de gradiente con diferenciación automática
#Parámetros del descenso de gradiente
alpha=0.9  #Tasa de aprendizaje
itera=10  #Número de iteraciones
w=np.random.rand() #Inicialización aleatoria de pesos

#Descenso de gradiente
def Descengad(x,w,t,alpha):
        gradient=funpcl(x,w,t)  #Calcular el gradiente
        return alpha*gradient

w_loss=[(w, loss(x,w,t))] #Registro de los valores de peso y perdida en un arreglo
for i in range(itera):
    w = w - Descengad(x,w,t,alpha)  #Actualización de pesos
    w_loss.append((w, loss(x,w,t)))  #Guardar el valor del peso y perdida en el arreglo

#Imprimir los pesos y perdidas por cada iteración
for i in range(0, len(w_loss)):
    print(f'w({i}): {w_loss[i][0]} \t loss: {w_loss[i][1]}')
```

```python
#Vizualización de las actualizaciones del gradiente
#Graficar función de perdida
plt.plot(ws, loss_ws,label='Perdida')
#Graficar pesos
for i in range(0, len(w_loss)-1):
    w1, c1 = w_loss[i]
    w2, c2 = w_loss[i+1]
    plt.plot(w1, c1, 'ro')
    plt.plot([w1, w2],[c1, c2], 'r-')
    plt.text(w1, c1+0.05, f'${i}$')
plt.plot(w2, c2, 'ro', label='$w(k)$')
plt.text(w2, c2+0.05, f'${i+1}$')
plt.xlabel('$w$', fontsize=12)
plt.ylabel('$\\xi$', fontsize=12)
plt.title('Actualizaciones trazadas en la función de pérdida')
plt.legend(loc=1)
plt.grid(True)
plt.show()
```

```python
#Graficar la línea ajustada contra la objetivo.
#Graficar los valores objetivos t contra las entradas x
plt.plot(x,t,'o',label='t')
#Graficar f(x)
ff=f(x)
plt.plot(x, ff, label='f(x)')
plt.plot([0,1],[0*w,1*w],label='y')
plt.xlabel('x',fontsize=12)
plt.ylabel('t',fontsize=12)
plt.title('Ajuste contra objetivo')
plt.legend(loc=2)
plt.grid(True)
plt.show()
```

# Conclusiones
Con el uso de la diferenciación automática se simplifica la implementación del descenso de gradiente en el entrenamiento de redes neuronales al eliminar la necesidad de calcular de forma manual los gradientes, así mismo se evitan posibles errores en el cálculo manual de gradientes. La diferenciación automática facilita el mantenimiento del código al reducir la cantidad de código manual necesario para calcular gradientes, lo que simplifica la depuración y la modificación del código a medida que se realizan mejoras en el modelo de la red neuronal.
En resumen, la implementación del descenso de gradiente con diferenciación automática en el entrenamiento de redes neuronales ofrece ventajas significativas en términos de simplificación, eficiencia, flexibilidad y facilidad de mantenimiento, lo que contribuye a un desarrollo más efectivo y robusto de modelos de aprendizaje automático.
