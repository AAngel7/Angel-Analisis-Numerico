Derivada
$$f(x+h)\approx f(x) + \frac{1}{1!}f'(x)h + \frac{1}{2!}f''(x)h^{2} + G(h)\\
f'(x)=\frac{f(x+h) - f(x)}{h} + G(h)$$
Diferencia pa'lante y pa'tras
$$f'(x)=\frac{f(x) - f(x-h)}{h} + G(h)$$
Fuentes de error.
Error de paroximación
$$\epsilon_{ap}=\frac{h}{2}|f''(x)|$$
Error aritmético
$$\epsilon_{ar}=\frac{2|f(x)|\epsilon_m}{h}$$
Error total
$$\epsilon_{total}=\epsilon_{ap} + \epsilon_{ar}$$
h optima
$$\sqrt{4\epsilon_m \frac{|f(X)}{|f''(X)|}}$$
